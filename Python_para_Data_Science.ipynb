#Python para Data Science

Data Science é uma combinação de várias disciplinas que une estatística, análise de dados e aprendizado de máquina (machine learning) para analisar dados além de extrair conhecimentos e conclusões destes dados.
É sobre coleta de dados, análise, tomada de decisões, encontrar padrões e realizar predições futuras.

AI-Venn-diagram.png

As empresas e instituições podem tomar decisões melhores, realizar previsões e descobrir padrões que nem sempre são óbvios na leitura de dados.

Alguns problemas que podem ser resolvidos:
- realizar previsões de quando um produto pode faltar em uma loja
- descobrir melhores rotas para logística
- prever atrasos em vôos
- criar promoções no varejo
- encontrar o melhor tempo para entregas
- prever o faturamento de uma empresa nos próximos anos
- prever eleições
- etc.

A Ciência de Dados pode ser aplicada em vários ramos onde existem dados:
- Indústria
- Comércio
- Política
- Logística
- E-commerce
- Bolsa de valores
- Estoques



## Cientista de Dados
Um cientista de dados deve ter os conhecimentos principalmente em machine learning, estatística, programação, matemática e banco de dados.

Como um cientista de dados trabalha?
- Realizar as perguntas corretas para entender o problema.
- Explorar e coletar os dados de alguma fonte.
- Extrair os dados, transformar os dados em um padrão.
- Limpar os dados para remover valores errados ou que podem interferir no resultado.
- Preencher lacunas para que a análise seja feita corretamente. As lacunas podem ser preenchidas com valores médios ou outro valor.
- Normalizar dados: os dados podem vir em diferentes escalas e isso deve ser padronizado. Por exemplo: em um registro pode ser encontrado o valor de 50 cm e em outro com o mesmo valor, porém registrado em 0,5m.
- Analisar os dados, encontrar padrões e realizar predições.
- Apresentar os resultados com conclusões e sugestões de maneira que outras pessoas entendam facilmente.



## Dados
Os dados são coleções de informações que podem se apresentar de forma estruturada ou não. Dados não estruturados devem ser organizados.
Por exemplo:
Pedro realizou uma pesquisa de satisfação de consumidores de uma loja. O método que foi utilizado foi a comunicação por mensagens instantâneas e e-mails. Pedro recebeu respostas de forma desestruturada, ou seja, ele deverá realizar a estruturação desses dados em uma planilha para obter os primeiros resultados.
Já Juliana realizou a mesma pesquisa, entretando ela enviou um link para os clientes, direcionando-os para um formulário eletrônico. A grande vantagem que ela obteve foi a de que o sistema de formulário eletrônico já entregava a ela uma planilha eletrônica com todos os dados estruturados.

Tipos de dados estuturados: listas, tabelas, bancos de dados...



## Python
Como trabalhar com Ciência de Dados utilizando Python?
É comum a utilização de algumas bibliotecas, como por exemplo:

- NumPy: é uma biblioteca matemática para trabalhar com vetores, álgebra linear, transformada de Fourier, etc.
- Pandas: é uma biblioteca para trablhar com dados estruturados. Com ela podemos importar arquivos CSV, ARFF, XLSX, além de criar dataframes - que são tabelas e realizar a preparação de dados.
- Matplotlib: é uma biblioteca para plotar gráficos a partir de dados.
- SciPy: é uma biblioteca de álgebra linear 
- Scikit-learn: é uma biblioteca de machine learning que contém algoritmos de classificação, regressão e agrupamento. Alguns algoritmos que incluem na biblioteca são SVM, Random Forest, Gradient Boosting, K-means, DBSCAN, etc.
- River: é uma biblioteca de machine learning para trabalhar com fluxo contínuo de dados e big data.

### Correlações
Mostra a correlação entre as colunas.

É uma tabela com números que representa como é o relacionamento de uma coluna com a outra. Os números variam de -1 a 1. O valor 1 representa uma correlação perfeita, ou seja, quando o valor de uma coluna sobe, a outra também acompanha. O valor -1 representa uma correlação inversa - quando uma sobe o valor, a outra desce. Valores próximos de 0 indicam baixo relacionamento entre as colunas.

##Exemplo com o dataset Iris
https://medium.com/@jebaseelanravi96/machine-learning-iris-classification-33aa18a4a983

https://towardsdatascience.com/exploring-classifiers-with-python-scikit-learn-iris-dataset-2bcb490d2e1b

https://github.com/terryz1/Iris_Classification/blob/master/iris_classification.ipynb

```python
from  sklearn import  datasets
iris=datasets.load_iris()

x=iris.data
y=iris.target

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.3)

from sklearn import tree
classifier = tree.DecisionTreeClassifier()
classifier.fit(x_train,y_train)
predictions=classifier.predict(x_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predictions))
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from pandas.plotting import parallel_coordinates
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
# load through url
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
attributes = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]
dataset = pd.read_csv(url, names = attributes)
dataset.columns = attributes
```

```python
dataset
```

```python
# types for the columns
dataset.dtypes
```

```python
dataset.describe()
```

```python
dataset.groupby('species').size()
```

```python
train, test = train_test_split(dataset, test_size = 0.4, stratify = dataset['species'], random_state = 42)
```

```python
train.groupby('species').size()
```

```python
n_bins = 10
fig, axs = plt.subplots(2, 2)
axs[0,0].hist(train['sepal_length'], bins = n_bins);
axs[0,0].set_title('Sepal Length');
axs[0,1].hist(train['sepal_width'], bins = n_bins);
axs[0,1].set_title('Sepal Width');
axs[1,0].hist(train['petal_length'], bins = n_bins);
axs[1,0].set_title('Petal Length');
axs[1,1].hist(train['petal_width'], bins = n_bins);
axs[1,1].set_title('Petal Width');

# add some spacing between subplots
fig.tight_layout(pad=1.0);
```

```python
sns.pairplot(train, hue="species", height = 2, palette = 'colorblind')
```

```python
corrmat = train.corr()
sns.heatmap(corrmat, annot = True, square = True);
```

```python
parallel_coordinates(train, "species", color = ['blue', 'red', 'green']);
```

```python
X_train = train[['sepal_length','sepal_width','petal_length','petal_width']]
y_train = train.species
X_test = test[['sepal_length','sepal_width','petal_length','petal_width']]
y_test = test.species
```

```python
mod_dt = DecisionTreeClassifier(max_depth = 3, random_state = 1)
mod_dt.fit(X_train,y_train)
prediction=mod_dt.predict(X_test)
print('The accuracy of the Decision Tree is',"{:.3f}".format(metrics.accuracy_score(prediction,y_test)))
```

```python
mod_dt.feature_importances_
```

```python
plt.figure(figsize = (10,8))
plot_tree(mod_dt, feature_names = ['sepal_length','sepal_width','petal_length','petal_width'],\
          class_names = train.species, filled = True)
```

```python
# Guassian Naive Bayes Classifier
mod_gnb_all = GaussianNB()
y_pred = mod_gnb_all.fit(X_train, y_train).predict(X_test)
print('The accuracy of the Guassian Naive Bayes Classifier on test data is',"{:.3f}".format(metrics.accuracy_score(y_pred,y_test)))
```

##Machine Learning



###Classificação
Processo de buscar e encontrar um modelo para separar dados em classes categóricas. Os dados são classificados em diferentes classes dependendo dos parâmetros e então rótulos podem ser previstos para os dados.

Exemplos de classificação: 
- binária: sim e não, classe 1 e classe 2, cachorro ou gato
- multi-classe: time a, b ou c; cachorro, gato ou rato.

Algoritmos de classificação em ML:
- Regressão Logística
- KNN
- SVN
- Naïve Bayes
- Árvores de decisão
- Random Forest

###Regressão
A regressão busca encontrar um modelo ou função para distinguir dados em valores reais contínuos ao invés de classficar utilizando rótulos.
Um algoritmo de regressão busca entrar uma função mapeadora para uma entrada para obter uma saída contínua.

Exemplos:
- Previsão do tempo
- Valor de uma ação em bolsa de valores
- etc.

Algoritmos de Regressão:
- Regressão linear simples
- Regressão linear múltipla
- Regressão Polinomial
- Regressão de Vetor de Suporte
- Regressão de árvore de decisão
- Regressão de Random Forest
